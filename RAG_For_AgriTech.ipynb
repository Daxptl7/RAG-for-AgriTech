```python
from langchain_community.document_loaders.pdf import PyPDFLoader
import os
```

```python
def load_all_pdfs():
  folder_path = "/content/AgriTech"
  num_doc = 0
  all_docs =[]

  for filename in os.listdir(folder_path):
    if filename.lower().endswith(".pdf"):
      pdf_path = os.path.join(folder_path, filename)
      pdf_loader = PyPDFLoader(pdf_path)
      docs = pdf_loader.load()
      all_docs.extend(docs)
      num_doc += 1

  print("The number of the PDFs:",num_doc)
  print("Total pages:",len(all_docs))
  return all_docs
```

```python
all_pdfs=load_all_pdfs()
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
def split_doc(documents , chunk_size = 500 , chunk_overlap=50):


  text_splitter = RecursiveCharacterTextSplitter(
      chunk_size=chunk_size,
      chunk_overlap=chunk_overlap
  )

  docs = text_splitter.split_documents(documents)
  return docs
```

```python
class EmbaddingManager:
  def __init__(self , model_name="all-MiniLM-L6-v2",):
    self.model_name=model_name
    print("Loding Model....",self.model_name)
    self.model = SentenceTransformer(model_name)
    print("Embadding diamentions:",self.model.get_sentence_embedding_dimension())

  def genearate_embadding(self , text):
    embadding = self.model.encode(text , show_progress_bar=True)
    return embadding
```

```python
import chromadb
import uuid
import os

class VectorStoreManager:

  def __init__(self , presist_directory="/content/Data/vector_store",collection_name="pdf_documents"):
    self.presist_directory = presist_directory
    self.collection_name = collection_name
    self.collection =None
    self.client = None
    self._initialize_store_()

  def _initialize_store_(self):
    os.makedirs(self.presist_directory , exist_ok=True)
    self.client = chromadb.PersistentClient(path=self.presist_directory)
    self.collection = self.client.get_or_create_collection(
        name=self.collection_name
        )

    print("Initialized Vector Store",self.collection_name)
    print("Document Collection:",self.collection.count())

  def add_docs(self , documents ,embaddings):
    if len(documents) != len(embaddings):
      raise ValueError("The number of documents and embaddings must be the same")


    max_batch_size = 4000 ,
    num_documents = len(documents)
    print(f"Attempting to add {num_documents} documents in batches.")

    for i in range(0, num_documents, max_batch_size):
      batch_docs = documents[i:i + max_batch_size]
      batch_embaddings = embaddings[i:i + max_batch_size]

      ids = []
      all_matadata = []
      document_content = []
      embadding_list = []

      for j , (doc , embadding) in enumerate(zip(batch_docs,batch_embaddings)):
        doc_id = f"doc_{uuid.uuid4()}"
        ids.append(doc_id)

        all_matadata.append(doc.metadata)
        document_content.append(doc.page_content)

        embadding_list.append(embadding.tolist())

      self.collection.add(
          ids=ids,
          metadatas=all_matadata,
          documents=document_content,
          embeddings=embadding_list
      )
      print(f"Added batch from index {i} to {min(i + max_batch_size, num_documents)}.")

    print("Document Collection:",self.collection.count())
    print("Total documents added in the collections:",len(documents))
```

```python
class RAGretrival:
  def __init__(self,embadding_managers,vector_store):
    self.vector_store = vector_store
    self.embadding_managers = embadding_managers # Corrected variable name


  def retrival(self , query , top_k=5 ,score_thresold=0):

    query_embadding = self.embadding_managers.genearate_embadding([query])
    query_embadding = query_embadding.tolist()

    result = self.vector_store.collection.query(
        query_embeddings=query_embadding,
        n_results=top_k,
        include=["documents","metadatas", "distances"] # Explicitly include distances
    )

    retrieval_docs=[]
    #cosine simalarity
    if result["documents"] and result["documents"][0]: # Corrected "document" to "documents"
      ids_list = result["ids"][0]
      documents_list = result["documents"][0]
      metadatas_list = result["metadatas"][0]
      distances_list = result["distances"][0] # Renamed for clarity and to avoid conflict

      for i , (doc_id , document_content ,metadata, distance_val)in enumerate(zip(ids_list,documents_list,metadatas_list,distances_list)):
        similarity_score = 1 - distance_val # Corrected variable name from 'distance' to 'distance_val'

        if(similarity_score > score_thresold):
          retrieval_docs.append( # Used .append() method for list
              {
              "id": doc_id, # Corrected semicolon to colon
              "page_content": document_content, # Corrected semicolon to colon
              "metadata": metadata, # Corrected semicolon to colon
              "score": similarity_score, # Corrected semicolon to colon
              "Rank": i+1 # Corrected semicolon to colon and added comma
              }
          )
      print(f"Retrived {len(retrieval_docs)} Documents")
      return retrieval_docs
```
